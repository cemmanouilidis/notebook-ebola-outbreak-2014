```python
%matplotlib inline

import sys
import re

import pandas as pd
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup
import urllib
```

```python
wiki = "http://en.wikipedia.org/wiki/2014_West_Africa_Ebola_outbreak"

permalink = "http://en.wikipedia.org/w/index.php?title=2014_West_Africa_Ebola_outbreak&oldid=620139889"

# download wiki article and pass it to BeautifulSoup
page = urllib.request.urlopen(permalink)
soup = BeautifulSoup(page)

# extract the 4th table in the permalink url
table = soup.findAll('table')[3]
```

```python
def cell_text(cell):
    return " ".join(cell.stripped_strings)

def get_table_columns(table):
    headers = []
    
    # get row containing the table headers
    row = table.find("tr")
    
    # skip first header "Date of report" this is the index for the DataFrame
    # squash sub headers into remaining main headers
    subheaders = ("cases", "deaths")
    for h in row.find_all("th")[1:]:
        for sh in subheaders:
            headers.append("%s %s" % (cell_text(h), sh))
     
    return headers

def get_table_data_and_index(table):
    data = []
    index = []
    
    # extract data for each row beginning with row #3
    # row #1 contains the headers 
    # row #2 contains subheaders "cases" and "deaths"
    for row in table.findAll('tr')[2:]:        
        rd = list(map(cell_text, row.find_all(re.compile('t[dh]') ))) 
        # row data for column "Date of report" go to the index
        index.append(rd[0])
        # remaining row data go into list data
        assert(len(rd[1:]) == 10)
        data.append(rd[1:])
        
    return (data, pd.DatetimeIndex(index))
```

```python
data, index = get_table_data_and_index(table)
columns = get_table_columns(table)

# create pandas DataFrame 
df = pd.DataFrame(data, index, columns, dtype=int)
df.sort(ascending=True, inplace=True)

df.head()
```

```python
# extract data for Guinea
dfg = pd.DataFrame(df, columns=("Guinea cases", "Guinea deaths"))

# sort firs, so we can calc the diff for cases and deaths
dfg.sort(ascending=True, inplace=True)

# create new column to calc new cases
dfg["Guinea cases new"] = (
    dfg["Guinea cases"] - dfg["Guinea cases"].shift()
).fillna(dfg["Guinea cases"]) # fill na value in first row with value of col cases

# create new column to calc new deaths
dfg["Guinea deaths new"] = (
    dfg["Guinea deaths"] - dfg["Guinea deaths"].shift()
).fillna(dfg["Guinea deaths"]) # fill na value in first row with value of col deaths

dfg.head()
```

```python
dfp = pd.DataFrame(dfg, columns=("Guinea cases new", "Guinea deaths new"))

ax = dfp.plot(figsize=(12,9))
ax.grid(False)
ax.set_frame_on(False)

ax.annotate(" -8", xy=("2014-06-24", -8))
ax.annotate(" -5", xy=("2014-07-08", -5))
ax.annotate(" -3", xy=("2014-07-15", -3))
ax.annotate(" -5", xy=("2014-07-15", -5))

ax.set_ylim(-12, 90)

ax.plot(
    ["Jan 01 2014", "Sep 01 2014"],
    [0, 0], 
    color='red', linestyle='dashed', linewidth=1
)

fig = ax.get_figure()
fig.savefig("static/ebola-outbreak-guinea-2014.png")

plt.show()
```
